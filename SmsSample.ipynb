# **Sentiment Analysis**

```python
import nltk
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
df["scores"]=df["message"].apply(lambda review: sid.polarity_scores(review))
df
```

```python
df["compound"] = df["scores"].apply(lambda x:x["compound"])
df
```

```python
df["comp_out"] = df["compound"].apply(lambda score : "ham" if score >=0.0 else "spam")
df.head(10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(df["label"],df["comp_out"])
```

# **NLP-Natural Language Processing**

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = [nlp(i) for i in df.message]
```

```python
df.head()
```

```python
message_new = [i.vector for i in doc]
message_new
```

# **Machine Learning**

```python
x= pd.concat([pd.DataFrame(message_new), df[['length', 'punct']]], axis=1)
x.columns = x.columns.astype(str)
y=df['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
from sklearn.svm import SVC
# Train the model on the training data
model = SVC().fit(x_train, y_train)
pred = model.predict(x_test)
```

# **Evaluation**

```python
# Machine Learning Evaluation
from sklearn.metrics import accuracy_score, classification_report

# Accuracy
accuracy = accuracy_score(y_test, pred)
print("Accuracy:", accuracy)

# Classification report
print("Classification Report:")
print(classification_report(y_test, pred))
```
